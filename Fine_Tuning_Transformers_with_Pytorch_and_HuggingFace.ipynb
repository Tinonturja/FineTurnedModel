## Dataset Preparation

```python
from datasets import load_dataset
dataset = load_dataset("yelp_review_full")

dataset
```

```python
dataset['train'] = dataset['train'].select([i for i in range(1000)])
dataset['test'] = dataset['test'].select([i for i in range(200)])
```

## Tokenizing Data

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')


# define a function to tokenize examples
def tokenize_function(examples):

  return tokenizer(examples["text"], padding="max_length", truncation=True)

# apply the function to the dataset
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
print(tokenized_datasets)
```

```python
#tokenized_datasets['train'][100]
```

```python
tokenized_datasets['train'][0].keys()
```

```python
# remove text columns from the dictionary
tokenized_datasets = tokenized_datasets.remove_columns(['text'])

# rename the label column to labels because the model expects the argument to be named labels
tokenized_datasets = tokenized_datasets.rename_column('label','labels')

tokenized_datasets.set_format(
    type="torch",
    columns=["input_ids", "token_type_ids", "attention_mask", "labels"]
)
```

```python
tokenized_datasets['train'][0]
```

## DataLoader

```python
from torch.utils.data import DataLoader
train_dataloader = DataLoader(
    tokenized_datasets["train"], shuffle=True, batch_size=2)

test_dataloader = DataLoader(
    tokenized_datasets["test"], batch_size=2
)
```

```python
sample_train_dataloader = next(iter(train_dataloader))
```

## Train the Model

```python
from transformers import AutoModelForSequenceClassification,BertForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-cased", num_labels=5)

# optimizer and learning rate
from torch.optim import AdamW
optimizer = AdamW(model.parameters(), lr=5e-5)

# set the number of epochs
num_epochs = 10


# calculate the total number of training steps
num_training_steps = num_epochs*len(train_dataloader)


# define the learning rate scheduler
from torch.optim.lr_scheduler import LambdaLR
lr_scheduler = LambdaLR(optimizer, lr_lambda = lambda current_step: 1 - (current_step/num_training_steps))
```

```python
import torch
# Check if CUDA is available and set the device accordingly
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

# Move the model to the appropriate device
model.to(device)
```

## Training the MOdel

```python
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
def train_model(model,tr_dataloader, val_dataloader):

  progress_bar = tqdm(range(num_training_steps))

  model.train()
  tr_losses = []

  for epoch in range(num_epochs):
    total_loss = 0

    # iterate over the training data batches
    for batch in tr_dataloader:

      # move the batches to the apropriate devices
      batch = {k: v.to(device) for k,v in batch.items()}

      # forward pass through the models
      outputs = model(**batch)

      # compute the loss
      loss = outputs.loss

      # backward pass
      loss.backward()

      total_loss += loss.item() # calculate the total loss for each epoch

      # update the model parameter
      optimizer.step()

      # update the learning rate scheduler
      lr_scheduler.step()

      # clear the gradients
      optimizer.zero_grad()

      # update the progressbar
      progress_bar.update(1)

    tr_losses.append(total_loss/len(tr_dataloader)) # total loss represents loss for one epochmnbvc

  plt.plot(tr_losses, label='Training Loss')
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.legend()
  plt.show()

```

## Evaluate

```python
from torchmetrics import Accuracy
```

```python
def evaluate(model, val_dataloader):

  model.eval()
  model.to(device)

  metric = Accuracy(task = 'multiclass', num_classes = 5).to(device)

  with torch.no_grad():

    for batch in val_dataloader:
      batch = {k:v.to(device) for k,v in batch.items()}


      outputs = model(**batch)
      logits = outputs.logits
      predictions = torch.argmax(logits, dim = -1)

      # accumulate the predictions and labels
      metric(predictions, batch['labels'])


  # compute the accuracy
  accuracy = metric.compute()
  print(f"Accuracy: {accuracy.item()}")
```

```python
# train model
train_model(model, train_dataloader, test_dataloader)
```

```python
# save the model
torch.save(model.state_dict(), 'customized_bert_model.pth')
```

```python
evaluate(model,test_dataloader)
```

## Training a conversational model using SFTTrainer

loading the timedettmers/openassistant-guanaco dataset from huggingface library

```python
assistant_dataset = load_dataset("timdettmers/openassistant-guanaco", split="train")
```

```python
print(assistant_dataset)
```

```python
assistant_dataset[100]['text']
```

```python
from trl import SFTTrainer,SFTConfig,DataCollatorForCompletionOnlyLM
from transformers import AutoModelForCausalLM,AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('facebook/opt-350m')

# create the model
model = AutoModelForCausalLM.from_pretrained('facebook/opt-350m')


## writing instruction and response template
instruction_template = '### Human:'
response_template = '### Assistant:'


# datacollator

collator = DataCollatorForCompletionOnlyLM(instruction_template=instruction_template,response_template=response_template,tokenizer=tokenizer, mlm=False)


training_args = SFTConfig(
    output_dir = "/tmp",
    num_train_epochs = 10,
    learning_rate = 2e-5,
    save_strategy = 'epoch',
    per_device_train_batch_size = 1,
    per_device_eval_batch_size = 1,
    max_seq_length = 512,
    do_eval = True
)

trainer = SFTTrainer(
    model = model,
    args = training_args,
    train_dataset = assistant_dataset,
    dataset_text_field = 'text',
    data_collator = collator
)
```

```python
from transformers import pipeline
pipe = pipeline("text-generation",model = model,tokenizer = tokenizer, max_new_tokens = 70)
print(pipe('''I have a budget lower than 30k USD. Which car I should buy? I prefer speed over comfort''')[0]['generated_text'])
```

```python
trainer.train()
```
